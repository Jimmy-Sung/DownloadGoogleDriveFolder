```python
import os
from google.colab import drive
from google.colab import files

# 1. 掛載 Google Drive (會跳出視窗要求授權)
drive.mount('/content/drive')

# ================= 設定區域 =================
# 請將這裡改成您要下載的資料夾路徑
# 建議去左側檔案欄找到該資料夾 -> 右鍵 -> 複製路徑
folder_path = '/content/drive/MyDrive/您的資料夾名稱'

# 設定壓縮後的檔名 (例如 output.zip)
output_filename = 'download.zip'
# ===========================================

# 2. 檢查路徑是否存在
if os.path.exists(folder_path):
    print(f"正在壓縮資料夾：{folder_path} ...")

    # 3. 使用 Linux 指令壓縮 (-r 代表遞迴包含子資料夾, -j 代表不包含層層路徑)
    # 這裡將資料夾壓縮到 /content/download.zip
    os.system(f'zip -r -j "/content/{output_filename}" "{folder_path}"')

    print("壓縮完成，準備下載...")

    # 4. 下載檔案到您的電腦
    files.download(f'/content/{output_filename}')
else:
    print("錯誤：找不到指定的資料夾路徑，請檢查名稱是否正確。")
```
